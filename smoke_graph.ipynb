```python
#%pip install seaborn
#%pip install pandas
#%pip install numpy
#%pip install matplotlib

import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
```

```python
data_smoke_drink = pd.read_csv("BDD_initial/addiction_population_data.csv", sep=",", index_col=0)
data_smoke_drink.head()
```

```python
ordre = ['Secondary', 'Primary', 'Postgraduate', 'University', 'College',
       'High School']
fig, ax = plt.subplots()
sns.violinplot(
    data=data_smoke_drink,
    x="education_level",
    y="sleep_hours",
    order=ordre,
    inner="quartile",
    ax=ax,
)
ax.set_title("Heures de sommeil par niveau académique")
ax.set_xlabel("Niveau académique")
ax.set_ylabel("Heures sommeils par nuit")
plt.show()

"""
Ce graphique en violon montre la distribution des heures de sommeil par nuit en fonction du niveau académique. 
Chaque "violon" représente la distribution des heures de sommeil pour un niveau académique spécifique, avec les quartiles indiqués à l'intérieur. 
On peut constater que le niveau académique ne semble pas avoir d'impact significatif sur le nombre d'heures de sommeil, car tous les quartiles des violons à plots se situent dans une plage similaire, avec une médiane aux alentours de 6,2 heures de sommeil.
"""
```

```python
fig = plt.figure()
ax = sns.barplot(data=data_smoke_drink, x="education_level", y="annual_income_usd", hue="addict_smoke", estimator=np.mean, ci=95)
ax.set_title("Revenu annuel moyen par niveau académique et addiction au tabac")
plt.show()

"""
Ce graphique à barres montre le revenu annuel moyen en fonction du niveau académique et de leurs addictions au tabac. 
On ne peut cependant pas visualiser une quelconque relation entre ces variables, le revenu annuel semble rester stable entre le niveau académique et l'addiction des personnes. 
"""
```

```python
fig = plt.figure()
ax = sns.barplot(data=data_smoke_drink, x="addict_smoke", y="has_health_issues", estimator=np.mean, hue="exercise_frequency")
ax.legend(loc="lower left", title="exercise_frequency")  
plt.show()

"""
Ce graphique nous montre une absence de relation entre la proportion de personnes ayant des problèmes de santé, leurs fréquences d'exercices physiques et si elles sont addictes au tabac. 
Cela est d'autant plus surprenant car il a été démontré que le tabac impacte négativement la santé. 
Nous expliquerons ce résultat par la suite. 
"""
```

```python
"""
Le lien entre l'addiction et les problèmes de santé étant surprenament inexistant, nous allons réaliser un test du chi² pour vérifier l'indépendance entre ces deux variables.
On prend alpha = 0.05 (seuil de fiabilité)
H0 = L'addiction au tabac et les problèmes de santé sont indépendants. 
H1 = L'addiction au tabac impactent les problèmes de santé.
"""
contingency_table = pd.crosstab(data_smoke_drink["addict_smoke"], data_smoke_drink["has_health_issues"])
print(contingency_table)

# test du chi²
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi² =", chi2)
print("p-value =", p, "> 0.05 donc on ne rejette pas H0")
print("Pour conclure, avec 5% de chance de se tromper, on peut affirmer que l'addiction au tabac et les problèmes de santé sont indépendants dans ce jeu de données.")
```
