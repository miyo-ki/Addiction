```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importation data bases
df1 = pd.read_csv("../BDD_initial/addiction_population_data.csv", sep=",", index_col=0)
df2 = pd.read_csv("../BDD_initial/mobile_addiction_data.csv", sep="\t", index_col=0)
df3 = pd.read_csv("../BDD_initial/Students Social Media Addiction.csv", sep=",", index_col=0)
```

analyses entre les variables

Matrice de corrélation de chaque data base

```python
#matrice de corrélation database 1, ici peut de corrélation positives et négative
plt.figure(figsize=(10,8))
corr = df1.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Matrice de corrélation")
plt.show()
```

```python
#matrice de corrélation database 2, ici peut de corrélation positives et négative
plt.figure(figsize=(10,8))
corr = df2.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Matrice de corrélation")
plt.show()
```

```python
#matrice de corrélation database 3, ici c'est la base la plus interessante car beaucoup de forte corrélations
plt.figure(figsize=(10,8))
corr = df3.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Matrice de corrélation")
plt.show()
```

Df2 - Mobile addiction

Addiction vs temps d'écran

```python
#droite de régression entre temps d'écran et addiction aux écrans
sns.scatterplot(data=df2, x='Daily_Screen_Time_Hours', y='Addiction_screen')
sns.regplot(data=df2, x='Daily_Screen_Time_Hours', y='Addiction_screen')
plt.show()
```

Addiction vs santé mentale

```python
#nuage de point entre santé mentale et addiction aux écrans
sns.scatterplot(data=df2, x='Mental_Health_Score', y='Addiction_screen')
plt.show()
```

Addiction vs activités physiques

```python
#nuage de point entre activité physique et addiction aux écrans
sns.scatterplot(data=df2, x='Physical_Activity_Hours', y='Addiction_screen')
plt.show()
```

Df3 - student social media addiction

Addiction vs screen time per day

```python
#nuage de point entre usage du téléphone et score d'addiction
sns.scatterplot(data=df3, x='Avg_Daily_Usage_Hours', y='Addicted_Score')
plt.show()
```

Addiction vs heures de sommeils

```python
#nuage de point entre heures de sommeil par nuit et score d'addiction
sns.scatterplot(data=df3, x='Sleep_Hours_Per_Night', y='Addicted_Score')
plt.show()
```

Addiction vs performances académique

```python
#nuage de point entre score d'addiction et performance académique(impact)
sns.scatterplot(data=df3, x='Affects_Academic_Performance', y='Addicted_Score')
plt.show()
```
