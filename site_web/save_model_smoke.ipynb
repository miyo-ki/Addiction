```python
import joblib
import os
import sys
import pandas as pd

sys.path.append('..')

from models.encoder import labelEncoder, oneHotEncoder
from models.random_forest_rgs_v2 import run as run_rf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Chargement des données
df = pd.read_csv('../BDD_initial/addiction_population_data.csv')

X_raw = df.drop(columns=['smokes_per_day'])
y     = df['smokes_per_day']

# OHE cette fois
X_ohe = oneHotEncoder(X_raw.copy())
X_train_ohe, X_test_ohe, y_train, y_test = train_test_split(X_ohe, y, test_size=0.2, random_state=42)

# Entraînement avec OHE
resultats_rf = run_rf(X_train_ohe, X_test_ohe, y_train, y_test)

# Encodeurs LabelEncoder pour le formulaire
cat_cols = df.select_dtypes(include='str').columns.tolist()
label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    le.fit(df[col].astype(str))
    label_encoders[col] = le

# Sauvegarde
os.makedirs('../site_web', exist_ok=True)
joblib.dump(resultats_rf['model'], '../site_web/model_smoke.pkl')
joblib.dump(label_encoders, '../site_web/encoders_smoke.pkl')
joblib.dump(list(X_ohe.columns), '../site_web/feature_names_smoke.pkl')

print("Modèle sauvegardé !")
```
